```python
import numpy as np
import matplotlib.pyplot as plt

# 📥 Load prices
prices = np.loadtxt("../prices.txt")
print("Shape:", prices.shape)      # Expect (50, N)

# 👀 Plot price history for a few instruments
plt.figure(figsize=(12, 6))
for i in range(5):  # Plot first 5 instruments
    plt.plot(prices[i], label=f"Instrument {i}")
plt.title("Price History of First 5 Instruments")
plt.xlabel("Day")
plt.ylabel("Price")
plt.legend()
plt.grid(True)
plt.show()
```
